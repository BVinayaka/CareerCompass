```python
import pandas as pd


unstructured_df['Experience'] = ""
unstructured_df['Keyword'] = ""


def extract_experience(description):
    
    if isinstance(description, str):
        
        if 'experience' in description.lower():
            
            experience_text = description.lower().split('experience')[0].strip()
            
            experience = ''.join(filter(str.isdigit, experience_text))
            return int(experience) if experience else 0
        else:
            return 0
    else:
        return 0

keywords = [
    'software', 'testing', 'teaching', 'support', 'analysis', 'management','business'
    'engineering', 'research', 'design', 'consulting', 'marketing', 'sales', 
    'finance', 'accounting', 'health', 'nursing', 'administration', 'legal', 
    'writing', 'editing', 'product', 'project', 'operations', 'customer', 
    'human resources', 'IT', 'software', 'hardware', 'network', 'security', 
    'data', 'business', 'logistics', 'supply chain', 'construction', 'real estate', 
    'manufacturing', 'quality', 'training', 'education', 'media', 'communications','assistant','trainee'
]


def extract_keyword(title):
   
    if isinstance(title, str):
        
        title_lower = title.lower()
        
        for keyword in keywords:
            if keyword in title_lower:
                return keyword.capitalize()
    return 'Other'


unstructured_df['Experience'] = unstructured_df['Description'].apply(extract_experience)
unstructured_df['Keyword'] = unstructured_df['Title'].apply(extract_keyword)


unstructured_df.to_csv('updated_job_dataset.csv', index=False)


unstructured_df.head()
```
