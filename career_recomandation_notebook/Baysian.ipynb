```python
import pandas as pd


df = pd.read_excel("dataset/correctness_results.xlsx")


def calculate_sum(group):
    
    return group.sum(axis=1)


result_dfs = []
column_names = []


for i in range(0, 89, 10):
    
    group = df.iloc[:, i:i+10]
    
    
    group_sum = group.sum(axis=1)
    
    
    column_name = f'Answer{i+1}-{i+10}'
    column_names.append(column_name)
    
    
    result_dfs.append(pd.DataFrame(group_sum, columns=[column_name]))


combined_result = pd.concat(result_dfs, axis=1)


career_option_mapping = {
    "Technology": 0,
    "Business": 1,
    "Healthcare": 2,
    "Teaching": 3,
    "Design": 4,
    "Engineering": 5,
    "Social": 6,
    "Hospital": 7,
    "Banking": 8
}


combined_result["Career Option"] = df["Career Option"].map(career_option_mapping)


combined_result.to_excel("dataset/summed_results_combined.xlsx", index=False)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import BayesianRidge
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd


df = pd.read_excel("dataset/summed_results_combined.xlsx")


label_map = {0: 'Technology', 1: 'Business', 2: 'Healthcare', 3: 'Teaching', 4: 'Design', 5: 'Engineering', 6: 'Social', 7: 'Hospital Nurse', 8: 'Banking'}


reverse_label_map = {v: k for k, v in label_map.items()}


X = df.drop("Career Option", axis=1)
y = df["Career Option"]


label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)


X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)


bayesian_ridge = BayesianRidge()
bayesian_ridge.fit(X_train, y_train)


last_row_predicted_label_bayesian = bayesian_ridge.predict(X_test.iloc[[-5]])[0]


predicted_career_option_bayesian = label_map.get(int(round(last_row_predicted_label_bayesian)), "Unknown")

print("\nPredicted Label for the Last Row in Testing Data (Bayesian Ridge):", round(last_row_predicted_label_bayesian))
print("Predicted Career Option for the Last Row in Testing Data (Bayesian Ridge):", predicted_career_option_bayesian)


y_pred_bayesian = bayesian_ridge.predict(X_test)


y_pred_bayesian = [int(round(label)) for label in y_pred_bayesian]

accuracy_bayesian = accuracy_score(y_test, y_pred_bayesian)
print("Accuracy of the Bayesian Ridge classifier:", accuracy_bayesian)
```
